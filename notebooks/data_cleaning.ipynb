```python
na_percent_per_col = df.isna().sum() * 100 / len(df)
na_percent_per_col = na_percent_per_col.to_frame().reset_index()
na_percent_per_col.columns = ['Variable', 'Percent NA']
na_percent_per_col
```

```python
#-128 to 127
cols_to_int8 = ['Severity']
cols_to_float8 = ['Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng', 'Humidity', 'Pressure', 'Visibility', 'Precipitation']
#-32768 to 32767
cols_to_float16 = ['Distance', 'Temperature', 'Wind_Chill', 'Wind_Speed']
```

```python
df_cleaned[cols_to_int8] = df_cleaned[cols_to_int8].astype(np.int8)
df_cleaned[cols_to_float8] = df_cleaned[cols_to_float8].astype(np.float32)
df_cleaned[cols_to_float16] = df_cleaned[cols_to_float16].astype(np.float32)
```

```python
df_cleaned.memory_usage() / 10**6
```
