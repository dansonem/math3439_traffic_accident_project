```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
data_path = '/content/drive/MyDrive/Data Science/Project 4/Project 4 Team 3/data/'
df = pd.read_csv(data_path + 'data_label_encoded.csv')
df.head()
```

#Severity dfs and graph font sizes

```python
df_sev1 = df[df['Severity'] == 1]
df_sev2 = df[df['Severity'] == 2]
df_sev3 = df[df['Severity'] == 3]
df_sev4 = df[df['Severity'] == 4]
```

```python
#graph font sizes from eve project 3
SMALL_SIZE = 20
MEDIUM_SIZE = 25
BIGGER_SIZE = 40

# Set font sizes individually. For example, you may want the title
# to be very big, but the axes labels to be only medium size.
plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SMALL_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

#Plot duration minutes for each severity

```python
col = 'Duration_Minutes'
plt.figure()
plt.boxplot([df_sev1[col], df_sev2[col], df_sev3[col], df_sev4[col]], showfliers=False)
plt.title(col + " (how long traffic was impacted)")
plt.show()
```

#Plot distance for each severity

```python
col = 'Distance'
plt.figure()
plt.boxplot([df_sev1[col], df_sev2[col], df_sev3[col], df_sev4[col]], showfliers=False)
plt.title(col + " (miles of road affected)")
plt.show()
```

#Plot description keywords for each severity

```python
col = 'Desc_Close'
plt.figure()
plt.hist([df_sev1[col], df_sev2[col], df_sev3[col], df_sev4[col]])
plt.title(col)
plt.show()
```
