#Changing weather condition options
Original dataset has 114 unique values for weather condition. There are some like mostly cloudy and partly cloudy that might be able to be combined into one 'cloudy' option, and some like volcanic ash that are rare and hard to classify. We want to simplify the possible values in weather condition.

```python
len(df['Weather_Condition'].unique())
```

```python
df['Weather_Condition'].value_counts().head(20)
```

```python
weather_and_counts = pd.DataFrame(df['Weather_Condition'].value_counts().reset_index())
weather_and_counts.columns = ['Weather_Condition', 'Value_Count']
weather_and_counts.head()
```

```python
weather_and_counts.to_csv(data_path + 'original_weather_conditions.csv', index=False)
```
