#Creating bool columns for presence of key words in description

```python
def find_keyword(keyword):
  found = df[df['Description'].str.lower().str.contains(keyword)]
  print('Number of df rows containing', keyword, 'in description: ', len(found))
  print('Percentage:', found.shape[0] / df.shape[0])
  found_sev1 = found[found['Severity'] == 1]
  print('Number of df sev1 rows containing', keyword, 'in description: ', len(found_sev1))
  print('Percentage:', found_sev1.shape[0] / df_sev1.shape[0])
  found_sev2 = found[found['Severity'] == 2]
  print('Number of df sev2 rows containing', keyword, 'in description: ', len(found_sev2))
  print('Percentage:', found_sev2.shape[0] / df_sev2.shape[0])
  found_sev3 = found[found['Severity'] == 3]
  print('Number of df  rows containing', keyword, 'in description: ', len(found_sev3))
  print('Percentage:', found_sev3.shape[0] / df_sev3.shape[0])
  found_sev4 = found[found['Severity'] == 4]
  print('Number of df rows containing', keyword, 'in description: ', len(found_sev4))
  print('Percentage:', found_sev4.shape[0] / df_sev4.shape[0])
```

#Label encoding - columns with only two options
- Side
- Sunrise_Sunset

```python
from sklearn.preprocessing import LabelEncoder
le_side = LabelEncoder()
le_sun = LabelEncoder()
side_transform = le_side.fit_transform(df['Side'])
sun_transform = le_sun.fit_transform(df['Sunrise_Sunset'])
label_map_side = dict(zip(df['Side'], side_transform)) # get the mapping between the original labels and encoded labels
label_map_sun = dict(zip(df['Sunrise_Sunset'], sun_transform)) # get the mapping between the original labels and encoded labels
print(label_map_side)
print(label_map_sun)
df['Side'] = side_transform
df['Sunrise_Sunset'] = sun_transform
```

#Date and time related columns
- Creating Duration_Minutes
- Extracting features from Start_Time and End_Time into new columns

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['End_Time'] = pd.to_datetime(df['End_Time'])
```

```python
duration_in_datetime = df['End_Time'] - df['Start_Time']
```

```python
duration_in_datetime.dt.components[['days', 'hours', 'minutes', 'seconds']]
```

```python
df['Duration_Minutes'] = duration_in_datetime.dt.components['days']*24*60 + duration_in_datetime.dt.components['hours']*60 + duration_in_datetime.dt.components['minutes']
df['Duration_Minutes'].head()
```

```python
df['Start_Time_Year'] = df['Start_Time'].apply(lambda time: time.year)
df['Start_Time_Month'] = df['Start_Time'].apply(lambda time: time.month)
df['Start_Time_Day'] = df['Start_Time'].apply(lambda time: time.day)
df['Start_Time_Hour'] = df['Start_Time'].apply(lambda time: time.hour)
df['Start_Time_Minute'] = df['Start_Time'].apply(lambda time: time.minute)

df['End_Time_Year'] = df['End_Time'].apply(lambda time: time.year)
df['End_Time_Month'] = df['End_Time'].apply(lambda time: time.month)
df['End_Time_Day'] = df['End_Time'].apply(lambda time: time.day)
df['End_Time_Hour'] = df['End_Time'].apply(lambda time: time.hour)
df['End_Time_Minute'] = df['End_Time'].apply(lambda time: time.minute)
```
